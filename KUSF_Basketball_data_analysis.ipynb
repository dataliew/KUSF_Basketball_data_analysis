```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_2022 = pd.read_excel('data/df_2022.xlsx')
df_2021 = pd.read_excel('data/df_2021.xlsx')
df_2020 = pd.read_excel('data/df_2020.xlsx')
```

```python
#Get sum() of the year from dataframes
stat_2022 = df_2022.sum()[['Scores','Assists','Rebounds','2-pointers','3-pointers']]
stat_2021 = df_2021.sum()[['Scores','Assists','Rebounds','2-pointers','3-pointers']]
stat_2020 = df_2020.sum()[['Scores','Assists','Rebounds','2-pointers','3-pointers']]
```

```python
labels = ['Score_sum', 'Assists_sum', 'Rebound_sum', '2pointer_sum', '3pointer_sum']

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, stat_2022 , width, label='2020')
rects2 = ax.bar(x, stat_2021, width, label='2021')
rects3 = ax.bar(x + width, stat_2020, width, label='2022')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Stats')
ax.set_title('KUSF Stat by years')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)

fig.tight_layout()

plt.show()
```

```python
#Save plot as an image file
import os
plt.savefig(os.path.join('Stat by years'), dpi=300, format='png', bbox_inches='tight')
```

```python
#Divide other columns by matches to generate average dataframes
df_2022_avg = df_2022.iloc[:,2:].div(df_2022['Matches'], axis=0)
df_2021_avg = df_2021.iloc[:,2:].div(df_2021['Matches'], axis=0)
df_2020_avg = df_2020.iloc[:,2:].div(df_2020['Matches'], axis=0)
```

```python
#Get average scores of each year
df_2022_avg_score = df_2022_avg['Scores']
df_2021_avg_score = df_2021_avg['Scores']
df_2020_avg_score = df_2020_avg['Scores']
df_avg_score_years = pd.DataFrame({'2020':df_2020_avg_score, '2021':df_2021_avg_score, '2022':df_2022_avg_score})
```

```python
df_avg_score_years.plot.density(figsize=(8,6), linewidth=4, fontsize=14)
plt.xlabel("Average Points")
```

```python
plt.savefig(os.path.join('Density of Average Scores of each year'), dpi=300, format='PNG', bbox_inches='tight')
```

```python
#Divide each columns by the number of matches to generate average dataframes
df_mid_avg = df_mid.iloc[:,2:].div(df_mid['Matches'], axis=0)
df_south_avg = df_south.iloc[:,2:].div(df_south['Matches'], axis=0)
df_mid_avg_scores = df_mid_avg['Scores']
df_south_avg_scores = df_south_avg['Scores']
df_avg_score_div = pd.DataFrame({'mid':df_mid_avg_scores, 'south':df_south_avg_scores})
```

```python
#Generate the dataframe of all the average records
df_all_avg = pd.concat([df_mid_avg, df_south_avg])
```

```python
corr = df_all_avg.corr()

fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,8,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(df_all_avg.columns)
ax.set_yticklabels(df_all_avg.columns)
plt.show()
```

```python
plt.savefig(os.path.join('Corrleation of Average Stats'), dpi=300, format='PNG', bbox_inches='tight')
```

```python
#Generate box plot of average scores
df_all_avg_scores = df_all_avg[['Scores']]
df_all_avg_scores.boxplot()
```

```python
#Get the value of upper bound to sort score outliers
q1 = np.quantile(df_all_avg_scores, 0.25)
q3 = np.quantile(df_all_avg_scores, 0.75)
med = np.quantile(df_all_avg_scores, 0.5)

IQR = q3 - q1
upper_bound = q3 + 1.5*IQR
```

```python
#Generate the dataframe sorted by score outliers
df_all_avg_score_outliers = df_all_avg.loc[df_all_avg['Scores'] >= upper_bound]
```

```python
corr_outlier = df_all_avg_score_outliers.corr()

fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr_outlier, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,8,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(df_all_avg_score_outliers.columns)
ax.set_yticklabels(df_all_avg_score_outliers.columns)
plt.show()
```
